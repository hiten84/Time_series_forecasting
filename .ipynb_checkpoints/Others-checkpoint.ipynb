```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('cash_flow.xlsb', engine='pyxlsb')
```

```python
df.head()
```

```python
df = df[['DT','OWNER','NET_DEMAND_ON','PAID_AMOUNT']]
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.columns
```

```python
df['PAID_AMOUNT'] = df['PAID_AMOUNT'].astype(float)
df['NET_DEMAND_ON'] = df['NET_DEMAND_ON'].astype(float)
```

```python
df = df.pivot_table(
    index=['DT', 'OWNER'],
    values = ['PAID_AMOUNT', 'NET_DEMAND_ON'],
    aggfunc = 'sum'
).reset_index()
```

```python
df.head(20)
```

```python
df.info()
```

```python
df.columns
```

```python
grouped_data = df.groupby(['DT', 'OWNER']).sum().reset_index()
grouped_data.set_index('DT', inplace=True)

grouped_data
```
